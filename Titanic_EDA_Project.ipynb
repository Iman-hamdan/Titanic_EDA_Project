```python
import pandas as pd

# Correct file path
file_path = '/Users/imanhamdan/Desktop/AI_projects/titanic3.xls'

# Load the dataset
df = pd.read_excel(file_path, engine='xlrd')  # Explicitly specify the engine

# Preview the data
print(df.head())  # Display the first 5 rows
print(df.info())  # Show data types and non-null counts
print(df.describe())  # Summary statistics for numerical columns
```

## Step 2: Check Missing Values
We examine the dataset for missing values. Identifying columns with missing data helps us decide how to handle them, either by filling, imputing, or dropping the affected rows or columns.


```python
# Check for missing values in the dataset
print(df.isnull().sum())
```

## Step 3: Univariate Analysis
Univariate analysis involves examining individual variables. We plot histograms to visualize the distribution of variables like age and survival count.


```python
df['age'].hist(bins=20, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
df['survived'].value_counts().plot(kind='bar')
plt.title('Survival Count')
plt.xlabel('Survived (1=Yes, 0=No)')
plt.ylabel('Frequency')
plt.show()
```

## Step 4: Bivariate Analysis
Bivariate analysis explores relationships between two variables. For example, we analyze survival rates based on passenger gender and class. We use bar plots and a heatmap to visualize these relationships and correlations.


```python
sns.barplot(data=df, x='sex', y='survived')
plt.title('Survival Rate by Gender')
plt.show()
```

```python
sns.barplot(data=df, x='pclass', y='survived')
plt.title('Survival Rate by Passenger Class')
plt.show()
```

```python
# Remove non-numeric columns from the dataframe
numeric_df = df.select_dtypes(include='number')

# Plot the correlation matrix
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

## Step 5: Handle Missing Data
We address missing data by:
1. Filling missing values in the `age` column with the median.
2. Dropping columns with excessive missing values, such as `cabin` and `boat`.


```python
print(df.isnull().sum())
```

```python
df['age'] = df['age'].fillna(df['age'].median())
```

```python
df = df.drop(columns=['cabin', 'boat', 'body'])  # Adjust columns as needed
```

## Step 6: Save the Cleaned Dataset
After cleaning the dataset, we save it to an Excel file for future use. This cleaned dataset excludes columns with excessive missing values and has imputed values where appropriate.


```python
df.to_excel('/Users/imanhamdan/Desktop/AI_projects/cleaned_titanic_data.xlsx', index=False)
print("Cleaned dataset saved!")
```
